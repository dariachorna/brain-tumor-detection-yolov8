# Аналіз датасету

```python
import yaml

with open(f'{dataset_path}/data.yaml', 'r') as f:
    print(f.read())
    data_yaml = yaml.safe_load(f)

class_names = data_yaml['names']
```

```python
for item in ['train', 'valid', 'test']:
    img_dir = f'{dataset_path}/{item}/images'
    count = len(os.listdir(img_dir))
    print(f'{item}: {count} images')
```

```python
class_counts = {}

for fold in ['train', 'valid', 'test']:
  lable_dir = f'{dataset_path}/{fold}/labels'

  for filename in os.listdir(lable_dir):
    lb = open(f'{lable_dir}/{filename}')

    for each_line in lb:

      parts = each_line.split()
      class_name = class_names[int(parts[0])]

      if class_name not in class_counts:
        class_counts[class_name] = 0
      class_counts[class_name] += 1

for name, count in class_counts.items():
  print(f'{name}: {count}')
```

Датасет має 3 класи:


*   glioma
*   meningioma
*   pituitary-tumor

Кількість зображень в обраному датасеті:


*   train: 1785 images
*   valid: 510 images
*   test: 254 images

Загальна кількість зображень: 2540

Баланс класів:


*   meningioma: 591
*   pituitary-tumor: 886
*   glioma: 1151


Можна зробити висновок, що присутній дисбаланс, але він не є критичним (клас glioma є більшим за клас meningioma у 1,94 рази).

Після перегляду вручну 20 зображень з різних класів зроблено висновок, що всі зображення мають чітке та щільне обрамлення зони, окрім двох випадків, де обрамлений фрагмент захоплював завелику частину зображення (на яку потрапляло забагато неураженої частини). Виділена зона чітко відрізняється структурою та кольором від іншої частини зображення, окрім одного випадку, де ураження є майже непомітним. Якість усіх продивлених зображень хороша.

```python
from ultralytics import YOLO
model = YOLO("yolov8n.pt")
results = model.train(data= "/content/brain-tumor-detection-yolo8-6/data.yaml", epochs = 50, imgsz = 640 )
```

Зазвичай для невеликих обсягів даних обирають 50-100 епох, тому я обрала мінімальне усереднене значення. Переглянувши файл results прийшла до висновку, що для обраного датасету було достатньо 25 епох, бо після 30-ої показники основних метрик не мали значного приросту, а надмірна кіллкість епох могла б (або, можливо, вже) призвести до запамʼятовування конкретних зображень, а не до тренування моделі.

# Оцінка результатів

Для аналізу метрик обрано три занчення: початкове та значення 25 і 50 епохи.


**mAP50** це середнє значення AP по всіх класах, яке в свою чергу є площою під кривою precision-recall, при Intersection over Union від 0,5. Ця метрика допомагає визначити якість пошуку та класифікації потрібного фрагмента на зображенні.


*   епоха 1: 0.21032
*   епоха 25: 0.6931
*   епоха 50: 0.73443

Стрімке зростання метрики спостерігається з 1 по 5 епоху, де показник зріс з 0.21032 до 0.50641.
Показник метрики зростає до 26-27 епохи, після чого
у діапазоні епох 28-50 зростання не спостергіається, а значення метрики коливються в межах 0.71 - 0.74.

**Precision** це метрика, що показує, яка частка з усіх передбачень моделі є правильною.

Так само, як і для mAP50 оберу три епохи для порівняння

*   епоха 1: 0.24451
*   епоха 25: 0.70334
*   епоха 50: 0.79617

Отже, стрімке зростання метрики можна спостерігати з 1 по 4 епоху, де показник зріс з 0.24451 до 0.64251.
Показник метрики зростає до 13 епохи включно, після чого
спостерігається лише коливання у межах 0.71 - 0.8.

**Recall** показує, яку частку реальних обʼєктів загалом модель знайшла, чим вищий цей показник, тим відповідно менше модель пропускатиме обʼєктів.

*   епоха 1: 0.1022
*   епоха 25: 0.67287
*   епоха 50: 0.72683

Стрімке зростання метрики спостерігається з 1 по 10 епоху, де показник зріс з 0.1022 до 0.61644.
Показник метрики зростає до 24 епохи, після чого
у діапазоні епох 28-50 спостергіаєтьсялише коливання у межах 0.68 - 0.72.
